```python
from nltk.corpus import *
```

```python
def preprocessing(texts):
    corpus = []
    stop_words  = set(stopwords.words('english'))
    punctuation = re.compile(r'[^a-zA-Z]+')
    count = 0
    for text in texts:
        try:
            count += 1
            lowered_text = [w.lower() for w in text if not re.match(punctuation, w)] 
            text_without_stopwords = [w for w in lowered_text if not w in stop_words]
            corpus.append(text_without_stopwords)
            if count%100 == 0:
                print "Preprocessed {} sentence".format(count)
        except:
            sys.exc_info()[0]
            next 
    print len(corpus)
    return corpus
```
